```python
a = cv.imread("road.jpeg")
b = cv.imread("road.jpeg")
c = cv.imread("road.jpeg")

a[:,:,2] = 0
cv.imshow("img",a)
cv.waitKey(0)



b[:,:,1] = 0
cv.imshow("img1",b)
cv.waitKey(0)

c[:,:,0] = 0
cv.imshow("img2",c)
cv.waitKey(0)
cv.destroyAllWindows()

```

```python
# Resmi yükle
resim_yolu = 'road.jpeg'
resim = cv.imread(resim_yolu)

# Resmin boyutlarını al
genislik, yukseklik = resim.shape[:2]

# Sağ tarafın belirli yüzdesini belirle
yuzde = -10  # Örnek olarak %70
kesilecek_genislik = int(genislik * (yuzde / 100))
kesilecek_baslangic = genislik - kesilecek_genislik

# Orijinal resmin devamını kopyala
devam_kisim = resim[:, kesilecek_genislik:].copy()

# Belirli bölgeyi seç
bolge = resim[:, kesilecek_baslangic:]

# Belirli filtre ve efektleri uygula
gray = cv.cvtColor(bolge, cv.COLOR_BGR2GRAY)
blurred = cv.GaussianBlur(gray, (5, 5), 0)
edges = cv.Canny(blurred, 50, 150)
cartoon = cv.bitwise_and(bolge, bolge, mask=edges)

# Orijinal resmin devamını ekleyerek tamamlayın
resim[:, kesilecek_genislik:] = devam_kisim

# Elde edilen efekti orijinal resmin üzerine yapıştır
resim[:, kesilecek_baslangic:] = cartoon

# Elde edilen resmi göster
cv.imshow('Elle Cizilmis Gorunum', resim)
cv.waitKey(0)
cv.destroyAllWindows()
```
